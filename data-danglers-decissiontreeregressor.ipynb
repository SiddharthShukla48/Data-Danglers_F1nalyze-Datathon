```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/f1nalyze-datathon-ieeecsmuj/train.csv")
```

**Feature Engineering**

```python
len(df)
```

```python
columns_to_drop = [ 'constructorRef','nationality','driverRef','number', 'positionText_x','time_x','year' ,'driver_code' ,'nationality_y','grand_prix','date','time_y','url_x','fp1_date','fp1_time','fp2_date','fp2_time','fp3_date','fp3_time','quali_date','quali_time','sprint_date','sprint_time','forename','surname','dob','url_y','positionText_y','company','url','fastestLapTime']
```

```python
identical_rows = df[df.duplicated(keep=False)]

# Count occurrences of each identical row
row_counts = identical_rows.value_counts()

print(row_counts)
```

```python
duplicates = df[df.duplicated(keep=False)]

# Count occurrences of each duplicate row
duplicate_counts = duplicates.groupby(list(duplicates.columns)).size().reset_index(name='count')

print(duplicate_counts)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['status'] = label_encoder.fit_transform(df['status'])
```

```python
# df.replace(r'\\N', np.nan, regex=True, inplace=True)

df.replace(r'\\N', np.nan, regex=True, inplace=True)

# Convert columns to numeric if necessary
df = df.apply(pd.to_numeric, errors='coerce')

# Fill NaN values with the mean of their respective columns
df.fillna(df.mean(), inplace=True)

# Verify that there are no more NaN values
print(df.isnull().sum())
```

**Loading Validation set**

```python
df_Validation = pd.read_csv("/kaggle/input/f1nalyze-datathon-ieeecsmuj/validation.csv")
```

```python
columns_to_drop_Validation = [ 'constructorRef','nationality','driverRef',         'number', 'positionText_x','time_x','year' ,'driver_code' ,'nationality_y','grand_prix','date','time_y','url_x','fp1_date','fp1_time','fp2_date','fp2_time','fp3_date','fp3_time','quali_date','quali_time','sprint_date','sprint_time','forename','surname','dob','url_y','positionText_y','company','url','fastestLapTime']

```

```python
df_Validation['status'] = label_encoder.fit_transform(df_Validation['status'])
```

```python
df_Validation = df_Validation.drop(columns=columns_to_drop_Validation)

df_Validation.replace(r'\\N', np.nan, regex=True, inplace=True)

# Convert columns to numeric if necessary
df_Validation = df_Validation.apply(pd.to_numeric, errors='coerce')

# Fill NaN values with the mean of their respective columns
df_Validation.fillna(df_Validation.mean(), inplace=True)

# Verify that there are no more NaN values
print(df_Validation.isnull().sum())

```

```python
df_Validation['status']
```

```python
df_encoded_Validation = df_Validation
```

**Saving Feature Engineered Validation Set**

```python
df_encoded_Validation.to_csv('/kaggle/working/full_processed_validation.csv', index=False)
```

**Training Model**

```python
from tqdm import tqdm 
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming your DataFrame is named 'df_encoded' and your validation DataFrame is named 'df_encoded_Validation'
X_train = df_encoded.drop(['position'], axis=1)  # Features
y_train = df_encoded['position']  # Target variable

X_val = df_encoded_Validation.drop(['position'], axis=1)  # Features
y_val = df_encoded_Validation['position']  # Target variable

# Initialize the model
model = DecisionTreeRegressor(max_depth=30, random_state=22,criterion= 'squared_error' )

# Fit the model
print("Fitting the model...")
model.fit(X_train, y_train)

# You can also use tqdm in cross-validation or if you are fitting the model in a loop
# For demonstration purposes, here we use tqdm with a range loop
print("Making predictions and evaluating the model...")
y_pred = model.predict(X_val)
print(y_pred)
print(y_val)

# Evaluate the model
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
print(f'RMSE: {rmse}')
```
